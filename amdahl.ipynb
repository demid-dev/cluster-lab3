```python
import matplotlib.pyplot as plt
import numpy as np

# Define the fraction of serial work
f_serial = 0.45

# Define the number of processors
n_processors = np.array([1, 2, 3, 4, 5, 6, 7, 8])

# Define the speedup values
speedup_actual = np.array([1, 1.47, 1.67, 1.67, 1.66, 1.65, 1.67, 1.67])

# Calculate the theoretical speedup based on Amdahl's Law
speedup_theoretical = 1 / (f_serial + ((1 - f_serial) / n_processors))

# Plot the actual and theoretical speedup graphs
plt.plot(n_processors, speedup_actual, 'ro-', label='Actual Speedup')
plt.plot(n_processors, speedup_theoretical, 'b*-', label='Theoretical Speedup')

# Set the plot title and axis labels
plt.title("Actual vs Theoretical Speedup Based on Amdahl's Law")
plt.xlabel("Number of Processors")
plt.ylabel("Speedup")

# Add a legend to the plot
plt.legend()

# Show the plot
plt.show()
```
